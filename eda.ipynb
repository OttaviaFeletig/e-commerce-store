```python
years = [2014, 2015, 2016, 2017]
```

```python
customer_acquisition = {"date": [], "year": [], "month": [],"total_customers":[], "unique_customer": [], "new_customers": [], "returning_customers": []}
```

```python
prev_month_customers = []

for year in years:
    for month in range(1, 13):
        customer_acquisition["date"].append(str(year) + "-" + str(month))
        customer_acquisition["year"].append(year)
        customer_acquisition["month"].append(month)
        customer_acquisition["total_customers"].append(len(df[(df["Order Year"] == year) & (
            df["Order Month"] == month)]["Customer ID"]))
        customer_acquisition["unique_customer"].append(len(df[(df["Order Year"] == year) & (
            df["Order Month"] == month)]["Customer ID"].unique()))
        # print(str(year) + " " + str(month) )
        if (year == 2014 and month == 1):

            customer_acquisition["new_customers"].append(0)
            customer_acquisition["returning_customers"].append(len(df[(df["Order Year"] == year) & (
                df["Order Month"] == month)]["Customer ID"].unique()))
            prev_month_customers = df[(df["Order Year"] == year) & (
                df["Order Month"] == month)]["Customer ID"].unique().tolist()

        else:
            current_month_customer = df[(df["Order Year"] == year) & (
                df["Order Month"] == month)]["Customer ID"].tolist()
            new_customer_this_month = 0
            returning_customers_this_month = 0
            for value in current_month_customer:
                if value not in prev_month_customers:
                    new_customer_this_month = new_customer_this_month + 1
                    # np.append(prev_month_customers, value)
                    prev_month_customers.append(value)
                else:
                    returning_customers_this_month = returning_customers_this_month + 1
            customer_acquisition["new_customers"].append(
                new_customer_this_month)
            customer_acquisition["returning_customers"].append(
                returning_customers_this_month)
   
```

```python
customer_acquisition_df = pd.DataFrame(customer_acquisition)
```
